Let's see what a workflow in Matplotlib would look like

```python
# Import and prepare the library
import matplotlib.pyplot as plt
%matplotlib inline

# Prepare the data
x = list(range(101))
y = []
for number in x:
  y.append(number ** 2)

# Prepare the plot area (fig) and the plot itself (ax) using plt.subplots()
fig, ax = plt.subplots()

# Add the data to the chart
ax.plot(x, y)

# Customize the plot by adding title to the plot and x and y axes
ax.set(title= "Acumulated earnings per day", xlabel= "Days", ylabel= "Earnings")

# Save our plot using fig.savefig()
fig.savefig("/graph-acumulated-revenue.png")
```

Now let's look at a scatter plot:

```python
# create a new dataset using the Numpy library
import numpy as np

x1 = np.linspace(0, 100, 20)
y1 = x1 ** 2

# create the x vs y scatter plot
fig, ax = plt.subplots()

ax.scatter(x1, y1)
```

```python
# We now visualize the sine function, using np.sin(X)
fig, ax = plt.subplots()

x2 = np.linspace(-10, 10, 100)
y2 = np.sin(x2)

ax.scatter(x2, y2)
```

Now let's look at another type of chart. For example, a bar chart, which usually associates numerical results with categorical variables (categories).

```python
# Let's create a dictionary with three dishes and their respective prices. 
# The keys of the dictionary will be the names of the meals, and the associated values, their price.
food = {"Pizza": 15, "Hamburguer": 11, "Sandwich": 17}
```

```python
# We will create a bar chart where the x-axis is formed by the dictionary keys,
# and the y-axis contains the values.
fig, ax = plt.subplots()
ax.bar(food.keys(), food.values())

# We add the corresponding titles
ax.set(title= "Food Prices", xlabel= "Food", ylabel= "Prices")
```

```python
# Let's try next with a horizontal bar chart.
fig, ax = plt.subplots()
ax.barh(list(food.keys()), list(food.values()))
ax.set(title= "Food Prices", ylabel= "Food", xlabel= "Prices")
```

Such a graph is a histogram. We can generate random numbers that follow a normal distribution (that accumulate around a central value), with the randn function:

```python
# We create a distribution of 1000 random values normally distributed
x = np.random.randn(1000)

# We create the histogram
fig, ax = plt.subplots()
ax.hist(x)
```

Now let's look at a more complex case, working with subplots, or figures containing several graphs:

```python
# We create a figure with 4 subgraphs (2 per row)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2, ncols= 2, figsize= (10, 5))
```

We add data to each of the graphs (axes)

```python
# We create the same chart layout, with a figure size of 10x5
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2, ncols= 2, figsize= (10, 5))

# For our first plot, we take the set x_1, y_1, and generate a line plot
ax1.plot(x1, y1)

# For our second chart, we take the set x_2, y_2, and generate a scatter chart
ax2.scatter(x2, y2)

# We create a graph with the prices of three meals in the lower left corner
ax3.bar(food.keys(), food.values())

# The graph in the lower right corner will be a histogram of random values with normal distribution
ax4.hist(np.random.randn(1000))
```

Having changed the style (the most obvious change we will see will be a grid at the background of each chart), we will also change the colors of the lines, dots and bars in each of the charts to hex codes of our preference:

```python
# Copiamos los valores de los gráficos anteriores
# We create the same chart layout, with a figure size of 10x5
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2, ncols= 2, figsize= (10, 5))

# For our first plot, we take the set x_1, y_1, and generate a line plot
ax1.plot(x1, y1, color= "#fcba03")

# For our second chart, we take the set x_2, y_2, and generate a scatter chart
ax2.scatter(x2, y2, color= "#fcba03")

# We create a graph with the prices of three meals in the lower left corner
ax3.bar(food.keys(), food.values(), color= "#03c6fc")

# The graph in the lower right corner will be a histogram of random values with normal distribution
ax4.hist(np.random.randn(1000), color= "#fc036b")
```
